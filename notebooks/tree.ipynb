```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import roc_auc_score, roc_curve, auc, accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA, NMF
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC


import sys
sys.path.append("../lib/")
from pathlib import Path
```

```python
dataset = "L_540_2022_C_E_R"
results_path = Path("../results/") / dataset
df = pd.read_csv(results_path / (dataset + "_preprocessed.csv"))
df = df.drop(columns=['file'], axis=1)

# Split the data into features and target
X = df.iloc[:, 1:] # Features
y = df.iloc[:, 0] # Target

# Perform label encoding on the categorical variables
le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])

# Split the data into features and target
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the different classifiers

# PCA-LDA
pca = PCA()
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

lda = LinearDiscriminantAnalysis()
lda.fit(X_train_pca, y_train)
y_pred_pca_lda = lda.predict(X_test_pca)

# NMF-LDA
nmf = NMF()
X_train_nmf = nmf.fit_transform(X_train)
X_test_nmf = nmf.transform(X_test)

lda = LinearDiscriminantAnalysis()
lda.fit(X_train_nmf, y_train)
y_pred_nmf_lda = lda.predict(X_test_nmf)

# FA-LDA
fa = LinearDiscriminantAnalysis()
fa.fit(X_train, y_train)
y_pred_fa_lda = fa.predict(X_test)

# Peak-LDA
X_train_peak = df.drop(['label'], axis=1).values
X_test_peak = X_test.values

lda = LinearDiscriminantAnalysis()
lda.fit(X_train_peak, y_train)
y_pred_peak_lda = lda.predict(X_test_peak)

# Logistic Regression with L1 regularization
logreg_l1 = LogisticRegression(penalty='l1')
logreg_l1.fit(X_train, y_train)
y_pred_logreg_l1 = logreg_l1.predict(X_test)

# Logistic Regression with L2 regularization
logreg_l2 = LogisticRegression(penalty='l2')
logreg_l2.fit(X_train, y_train)
y_pred_logreg_l2 = logreg_l2.predict(X_test)




knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)

```

```python
# Plot ROC curves for each classifier
plt.figure(figsize=(10, 8))
for clf in [knn, dt, nb, rf, gb]:
    y_prob = clf.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob[:, 1], pos_label=1)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label='%s ROC (AUC = %0.2f)' % (clf.__class__.__name__, roc_auc))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves of Different Classifiers')
plt.legend(loc="lower right")
plt.show()
```

```python
# Calculate the accuracy scores
acc_knn = accuracy_score(y_test, y_pred_knn)
acc_dt = accuracy_score(y_test, y_pred_dt)
acc_nb = accuracy_score(y_test, y_pred_nb)
acc_rf = accuracy_score(y_test, y_pred_rf)
acc_gb = accuracy_score(y_test, y_pred_gb)

# Plot the accuracy scores
classifiers = ['KNN', 'Decision Tree', 'Naive Bayes', 'Random Forest', 'Gradient Boosting']
accuracies = [acc_knn, acc_dt, acc_nb, acc_rf, acc_gb]


plt.bar(classifiers, accuracies)
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Accuracy of different classifiers')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(accuracies, labels=classifiers)
ax.set_xlabel('Classifier')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of different classifiers')
plt.show()
```

```python
# Define the models
models = [
    KNeighborsClassifier(n_neighbors=5),
    DecisionTreeClassifier(random_state=0),
    GaussianNB(),
    RandomForestClassifier(random_state=0),
    GradientBoostingClassifier(random_state=0),
]

# Define the names of the models
names = [
    "k-Nearest Neighbors",
    "Decision Trees",
    "Naive Bayes",
    "Random Forest",
    "Gradient Boosting",
]
```

```python
# Perform cross-validation for each model
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
scores = []
for model in models:
    score = cross_val_score(model, X, y, cv=cv, scoring="accuracy")
    scores.append(score)
```
